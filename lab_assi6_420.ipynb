```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Q1: Mathematical Function Plot
M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 100)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label=r'$y = M \cdot x^2$', color='b', linestyle='--')
plt.plot(x, y2, label=r'$y = M \cdot \sin(x)$', color='r', linestyle='-')

plt.title("Mathematical Function Plots")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid()
plt.show()


# Q2: Seaborn Bar Plot of Subject Scores
subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'Computer Science']
scores = [85, 90, 78, 88, 92]

df = pd.DataFrame({'Subject': subjects, 'Score': scores})

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subject', y='Score', data=df, palette='coolwarm')

# Annotate bars with values
for p in ax.patches:
    ax.annotate(f"{p.get_height()}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=12, color='black')

plt.title("Subject Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# Q3: Random Dataset with Roll Number Seed
roll_number = 420
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot: Cumulative Sum
axs[0, 0].plot(np.cumsum(data), color='blue')
axs[0, 0].set_title("Cumulative Sum Plot")
axs[0, 0].set_xlabel("Index")
axs[0, 0].set_ylabel("Cumulative Sum")
axs[0, 0].grid()

# Scatter Plot: Random Noise
axs[0, 1].scatter(range(50), data, color='red')
axs[0, 1].set_title("Scatter Plot with Noise")
axs[0, 1].set_xlabel("Index")
axs[0, 1].set_ylabel("Random Values")
axs[0, 1].grid()

# Hide the remaining empty subplots
axs[1, 0].axis('off')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()


# Q4: Read & Analyze Company Sales Data
url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
sales_df = pd.read_csv(url)

# 1. Line Plot: Total Profit Over Months
plt.figure(figsize=(8, 5))
sns.lineplot(x='month_number', y='total_profit', data=sales_df, marker='o', color='g')
plt.title("Total Profit Over Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.grid()
plt.show()

# 2. Multiline Plot for Product Sales
plt.figure(figsize=(10, 6))
for column in sales_df.columns[1:-1]:  # Exclude month_number and total_profit
    plt.plot(sales_df['month_number'], sales_df[column], marker='o', label=column)

plt.title("Product Sales Over Months")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()

# 3. Bar Chart for All Features
sales_df.plot(kind='bar', figsize=(12, 6))
plt.title("Sales Data Bar Chart")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend()
plt.grid()
plt.show()
```
