```python
import numpy as np
import matplotlib.pyplot as plt

# Q1: Matrix Operations
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

# i. Sum of all elements
sum_all = np.sum(gfg)

# ii. Sum of all elements row-wise
sum_rows = np.sum(gfg, axis=1)

# iii. Sum of all elements column-wise
sum_cols = np.sum(gfg, axis=0)

print("Q1:")
print("Sum of all elements:", sum_all)
print("Sum of elements row-wise:", sum_rows)
print("Sum of elements column-wise:", sum_cols)

# Q2: Array Operations
array1 = np.array([10, 52, 62, 16, 16, 54, 453])

# i. Sorted array
sorted_array = np.sort(array1)

# ii. Indices of sorted array
sorted_indices = np.argsort(array1)

# iii. 4 smallest elements
smallest_4 = np.sort(array1)[:4]

# iv. 5 largest elements
largest_5 = np.sort(array1)[-5:]

print("\nQ2(a):")
print("Sorted Array:", sorted_array)
print("Indices of sorted array:", sorted_indices)
print("4 Smallest elements:", smallest_4)
print("5 Largest elements:", largest_5)

# Q2(b): Integer and Float elements
array2 = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
integer_elements = array2[array2 == np.floor(array2)]
float_elements = array2[array2 != np.floor(array2)]

print("\nQ2(b):")
print("Integer elements:", integer_elements)
print("Float elements:", float_elements)

# Q3: Weekly Sales Dataset
initials_ascii_sum = ord('B') + ord('R')
X = initials_ascii_sum

sales = np.array([X, X+50, X+100, X+150, X+200])

# b) Compute tax rate
tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales * (1 + tax_rate)

# c) Apply discount
discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)

# d) Expand sales data for 3 weeks
weekly_sales = np.vstack([sales, sales * 1.02, sales * 1.04])  # 2% increase per week

print("\nQ3:")
print("Sales Data:", sales)
print("Tax Rate:", tax_rate)
print("Sales after tax:", taxed_sales)
print("Sales after discount:", discounted_sales)
print("Weekly Sales:\n", weekly_sales)

# Q4: Generate x values and compute y
x = np.linspace(-10, 10, 100)

y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(x, y1, label='y = x^2', color='b')
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='y = sin(x)', color='g')
plt.title('y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='y = e^x', color='r')
plt.title('y = e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='y = log(|x|+1)', color='m')
plt.title('y = log(|x|+1)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.tight_layout()
plt.show()
```
