Q.6: Employee Dataset and Perform Operations

```python
# Create dataset for employees
employee_data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}

employee_df = pd.DataFrame(employee_data)

# Save to CSV (if required)
employee_df.to_csv("employees.csv", index=False)
```

a) Shape of the DataFrame:

```python
print(employee_df.shape)
```

b) Summary of the DataFrame:

```python
print(employee_df.info())
```

c) Generate Descriptive Statistics:

```python
print(employee_df.describe())
```

d) First 5 and Last 3 Rows:

```python
print(employee_df.head())


print(employee_df.tail(3))
```

e) Calculate Specific Statistics:

```python
avg_salary = employee_df['Salary'].mean()
print("Average Salary:", avg_salary)


total_bonus = employee_df['Bonus'].sum()
print("Total Bonus Paid:", total_bonus)


youngest_age = employee_df['Age'].min()
print("Youngest Employee's Age:", youngest_age)


highest_rating = employee_df['Rating'].max()
print("Highest Performance Rating:", highest_rating)
```

f) Sort by Salary in Descending Order:

```python
sorted_df = employee_df.sort_values(by='Salary', ascending=False)
print(sorted_df)
```

g) Add a New Column for Performance Rating Category:

```python
def categorize_performance(rating):
    if rating >= 4.5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'

employee_df['Performance_Category'] = employee_df['Rating'].apply(categorize_performance)
print(employee_df)
```

h) Identify Missing Values:

```python
print(employee_df.isnull().sum())
```

i) Rename Employee_ID to ID:

```python
employee_df.rename(columns={'Employee_ID': 'ID'}, inplace=True)
print(employee_df)
```

j) Filter Employees with Specific Criteria:

```python
experienced_employees = employee_df[employee_df['Years_of_Experience'] > 5]
print(experienced_employees)


it_employees = employee_df[employee_df['Department'] == 'IT']
print(it_employees)
```

k) Add a New Column for Tax Deduction:

```python
employee_df['Tax'] = employee_df['Salary'] * 0.10
print(employee_df)
```
