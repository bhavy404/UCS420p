```python
### **Lab Assignment 7 - UCS420 Cognitive Computing**
### **Name: Bhavya Raj Singh**
### **Roll Number: 102497003**

# **Q1: Sales Data Analysis**

## **Part I: Randomized Sales Data Generation**

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize Random Seed
roll_number = 102497003  # Use an integer instead of a list
np.random.seed(roll_number % 1000)  # Seed should be an integer

# Generate Sales Data (12 months, 4 categories)
sales_data = np.random.randint(1000, 5000, size=(12, 4))

# Convert to Pandas DataFrame
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
df = pd.DataFrame(sales_data, columns=categories, index=months)

# Display First 5 Rows
print("Initial Sales Data:")
print(df.head())

## **Part II: Data Manipulation and Analysis**

# Summary Statistics
print("\nSummary Statistics:")
print(df.describe())

# Total Sales per Category and Month
df["Total Sales"] = df.sum(axis=1)
print("\nTotal Sales per Month:")
print(df["Total Sales"])

# Growth Rate Calculation
df["Growth Rate"] = df["Total Sales"].pct_change() * 100
print("\nSales Data with Growth Rate:")
print(df)

# Apply Discount Based on Roll Number
if roll_number % 2 == 0:
    df["Electronics"] = (df["Electronics"] * 0.9).astype(int)  # 10% discount on Electronics
else:
    df["Clothing"] = (df["Clothing"] * 0.85).astype(int)  # 15% discount on Clothing

print("\nSales Data After Discount:")
print(df)

## **Part III: Visualizations**

# Line Plot for Monthly Sales Trends
plt.figure(figsize=(10, 5))
for category in categories:
    plt.plot(df.index, df[category], marker='o', label=category)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trends")
plt.legend()
plt.grid()
plt.show()

# Box Plot for Sales Distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution Across Categories")
plt.show()

# **Q2: NumPy Array Operations**

array = np.array([[1, -2, 3], [-4, 5, -6]])

# Element-wise Absolute Values
print("Element-wise Absolute Values:\n", np.abs(array))

# Percentiles
flattened = array.flatten()
print("\nPercentiles (Flattened): 25th:", np.percentile(flattened, 25), "50th:", np.percentile(flattened, 50), "75th:", np.percentile(flattened, 75))

# Mean, Median, Standard Deviation
print("\nMean:", np.mean(flattened))
print("Median:", np.median(flattened))
print("Standard Deviation:", np.std(flattened))

# **Q3: NumPy Rounding Operations**

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

print("Floor:", np.floor(a))
print("Ceil:", np.ceil(a))
print("Truncated:", np.trunc(a))
print("Rounded:", np.round(a))

# **Q4: Swap Two Elements in a List**

arr = [1, 2, 3, 4, 5]
print("Original List:", arr)
arr[1], arr[3] = arr[3], arr[1]
print("List After Swap:", arr)

# **Q5: Swap Two Elements in a Set**

s = {10, 20, 30, 40}
print("Original Set:", s)

s_list = list(s)
s_list[0], s_list[2] = s_list[2], s_list[0]
s = set(s_list)

print("Set After Swap:", s)
```
